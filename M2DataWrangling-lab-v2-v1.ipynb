### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# View unique entries in the 'Country' column
unique_countries = df['Country'].unique()
print(f"Number of unique entries in 'Country': {len(unique_countries)}")
print(unique_countries)
```

```python
# Count occurrences of each country
country_counts = df['Country'].value_counts()

# Flag entries with 2 respondents
inconsistent_countries = country_counts[country_counts <= 2]
print("Potential inconsistent or irrelevant entries:\n", inconsistent_countries)
```

```python
country_mapping = {
    "Democratic People's Republic of Korea": "North Korea",
    "Republic of Korea": "South Korea",
    "United States of America": "USA",
    "Russian Federation": "Russia",
    "Venezuela, Bolivarian Republic of...": "Venezuela",
    "Iran, Islamic Republic of...": "Iran"
}
df['Country'] = df['Country'].replace(country_mapping)
```
